```python
from telebot.types import Message

@bot.message_handler(commands=['moviehelp'])
def help(message: Message):
    help_message = (
         "Hello there!🎬 Welcome to the world of cinema! Explore movie magic with these commands: 🍿\n\n"
        "You can access the movie-related features using the following commands:\n\n"
        "1. /popularity :Discover the top 10 trending movies! 🌟\n"
        "2. /topmovie : Find the cream of the crop - the top-rated movie! 🏆\n"
        "3. /moviedetails [movie_title] :Delve into the details of any movie. Remember to enter the movie title accurately for precise results! 🔍🎥\n\n"
    )
    bot.reply_to(message, help_message)

bot.polling()
```

```python
# Command handler function for /moviedetails command
@bot.message_handler(commands=['moviedetails'])
def movie_details(message):
    title = message.text[14:]
    movie = df[df['title'].str.contains(title, case=False, na=False)].iloc[0]
    if not movie.empty:
        movie_details_message = f"Movie Details for '{title}':\n\n" \
                                f"Title: {movie['title']}\n" \
                                f"Release Date: {movie['release_date']}\n" \
                                f"Overview: {movie['overview']}\n" \
                                f"Rating: {movie['vote_average']}\n" \
                                f"Vote Count: {movie['vote_count']}\n"
        bot.reply_to(message, movie_details_message)
    else:
        bot.reply_to(message, f"No movie found with the title '{title}'")

# Command handler function for /popularmovies command
@bot.message_handler(commands=['popularity'])
def popular_movies(message):
    popular_movies_df = df.nlargest(10, 'vote_count')
    movie_details_message = "Top 10 Popular Movies:\n\n"
    for index, movie in popular_movies_df.iterrows():
        movie_details_message += f"Title: {movie['title']}\n" \
                                  f"Release Date: {movie['release_date']}\n" \
                                  f"Overview: {movie['overview']}\n" \
                                  f"Rating: {movie['vote_average']}\n" \
                                  f"Vote Count: {movie['vote_count']}\n\n"
    bot.reply_to(message, movie_details_message)

# Command handler function for /topmovie command
@bot.message_handler(commands=['topmovie'])
def top_movies(message):
    top_movie_details = df.iloc[0]
    movie_details_message = f"Top Rated Movie Details:\nTitle: {top_movie_details['title']}\n" \
                            f"Overview: {top_movie_details['overview']}\n" \
                            f"Release Date: {top_movie_details['release_date']}\n" \
                            f"Rating: {top_movie_details['vote_average']}\n" \
                            f"Vote Count: {top_movie_details['vote_count']}"
    bot.reply_to(message, movie_details_message)

bot.polling()
```
